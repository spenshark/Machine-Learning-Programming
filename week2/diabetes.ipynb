```python
# diabetes.csv 데이터셋으로 실습

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# -----------------------------
# 1) 데이터 준비
# -----------------------------
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/machine-learning-practice/Machine-Learning-Programming/week2/data/diabetes.csv", header=0).dropna()

X = df.drop(columns=["Outcome"])
y = df["Outcome"]

# 여기부터는 데이터셋이 달라져도 코드 변경없이 그대로 가져다 쓰면 됨
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# -----------------------------
# 2) 모델 구성
# -----------------------------

# 모델 만들기
dt = DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(n_estimators=200, random_state=42)
lr = LogisticRegression(max_iter=500)
```

```python
# -----------------------------
# 3) 모델 학습
# -----------------------------

# fit 함수를 통해 학습
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
lr.fit(X_train, y_train)
```

```python
# -----------------------------
# 4) 모델 평가
# -----------------------------

# 예측, predict함수 -> 테스트 할 데이터를 인자로 전달
y_pred = dt.predict(X_test)
dt_acc = accuracy_score(y_test, y_pred)

#dt_acc = accuracy_score(y_test, dt.predict(X_test))
rf_acc = accuracy_score(y_test, rf.predict(X_test))
lr_acc = accuracy_score(y_test, lr.predict(X_test))
```

```python
# -----------------------------
# 5) 결과 출력
# -----------------------------

print("=== Test Accuracy ===")
print(f"Decision Tree : {dt_acc:.4f}")
print(f"Random Forest : {rf_acc:.4f}")
print(f"Logistic Reg. : {lr_acc:.4f}")
```
